```
# imports.
import pandas as pd 
import datetime as dt
from google.colab import files
```

```
# read in the csv files.
df1 = pd.read_csv('bodystats_garmin.csv', skiprows=1)
df2 = pd.read_csv('activitystats_garmin.csv', skiprows=2, thousands=',')
df3 = pd.read_csv('fitbit_export_body.csv', skiprows=1)
df4 = pd.read_csv('fitbit_export_activities.csv', skiprows=1)
df5 = pd.read_csv('fitbit_export_sleep.csv', skiprows=1)
```

# Body.

```
# convert to datetime.
df1['Date'] = pd.to_datetime(df1['Date'])
# split date into new columns.
df1['Day'] = df1['Date'].dt.day
df1['Month'] = df1['Date'].dt.month
df1['Year'] = df1['Date'].dt.year
# change the data type of weight, fat, date.
df1['Weight'] = df1['Weight'].astype(int)
df1['Fat'] = df1['Fat'].astype(int)
df1['Date'] = df1['Date'].astype(str)
# single out each year.
df1_2020 = df1['Year']==2020
df1_2020 = df1[df1_2020]
df1_2020 = df1_2020.drop(columns=['Day', 'Month', 'Year'])
df1_2019 = df1['Year']==2019
df1_2019 = df1[df1_2019]
df1_2019 = df1_2019.drop(columns=['Day', 'Month', 'Year'])
df1_2018 = df1['Year']==2018
df1_2018 = df1[df1_2018]
df1_2018 = df1_2018.drop(columns=['Day', 'Month', 'Year'])
df1_2017 = df1['Year']==2017
df1_2017 = df1[df1_2017]
df1_2017 = df1_2017.drop(columns=['Day', 'Month', 'Year'])
df1_2016 = df1['Year']==2016
df1_2016 = df1[df1_2016]
df1_2016 = df1_2016.drop(columns=['Day', 'Month', 'Year'])
df1_2015 = df1['Year']==2015
df1_2015= df1[df1_2015]
df1_2015 = df1_2015.drop(columns=['Day', 'Month', 'Year'])
```
